```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
ds = pd.read_csv(r'C:\Users\Shashank\Desktop\titanic.csv')
print(ds.columns)
print(len(ds.columns))
ds.head()
```

```python
cols_to_remove = [
    'PassengerId',
    'Name',
    'Cabin',
    'Embarked',
    'Ticket'
]
df = ds.drop(cols_to_remove, axis=1)
print(len(df.columns))

df.head()
```

```python
sex_mapping = {
    'male': 0,
    'female': 1,
}
df.Sex = df.Sex.map(sex_mapping)
```

```python
data = df.values
print(data.shape)

df = df.dropna(axis=0)

data = df.values
print(data.shape)
```

```python
X = data[:,1:] ## input
Y = data[:, 0] ## labels

print(X.shape)
print(Y.shape)

split = int(0.8*X.shape[0])

X_train = X[:split]
Y_train = Y[:split]

X_test = X[split:]
Y_test = Y[split:]

print("-----------")

print(X_train.shape)
print(X_test.shape)

print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# all parameters not specified are set to their defaults
logisticRegr = LogisticRegression()
```

```python
logisticRegr.fit(X_train, Y_train)
```

```python
score = logisticRegr.score(X_test, Y_test)
print(score)
```
